```python
import pandas as pd
import matplotlib.pyplot as plt

# Chemin du rapport de drive
drift_report_path = "./data/monitoring/drift_report.csv"

# Charger le rapport
drift_report = pd.read_csv(drift_report_path)

# Afficher les premires lignes du rapport
print(drift_report.head())

# Visualisation
plt.figure(figsize=(12, 6))
plt.bar(drift_report['feature'], drift_report['statistic'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Feature')
plt.ylabel('Drift Statistic')
plt.title('Drift Detection Statistics per Feature')
plt.tight_layout()
plt.show()
```
